```python
trans = [
    transitions.OpticalTransition(
        transitions.OpticalTransitionType.R, J_ground=0, F1=3 / 2, F=1
    ),
    transitions.OpticalTransition(
        transitions.OpticalTransitionType.P, J_ground=2, F1=3 / 2, F=1
    )
]

polarizations = [[couplings.polarization_Z], [couplings.polarization_Z]]

transition_selectors = couplings.generate_transition_selectors(
    trans,
    polarizations,
    ground_mains=[
        1
        * states.CoupledBasisState(
            J=0,
            F=0,
            F1=1 / 2,
            mF=0,
            I1=1 / 2,
            I2=1 / 2,
            Ω=0,
            P=+1,
            electronic_state=states.ElectronicState.X,
        ),
        1
        * states.CoupledBasisState(
            J=2,
            F=2,
            F1=5 / 2,
            mF=0,
            I1=1 / 2,
            I2=1 / 2,
            Ω=0,
            P=+1,
            electronic_state=states.ElectronicState.X,
        )
    ],
    excited_mains=[
        1
        * states.CoupledBasisState(
            J=1,
            F=1,
            F1=3 / 2,
            mF=0,
            I1=1 / 2,
            I2=1 / 2,
            Ω=1,
            P=-1,
            electronic_state=states.ElectronicState.B,
        ),
        1
        * states.CoupledBasisState(
            J=1,
            F=1,
            F1=3 / 2,
            mF=0,
            I1=1 / 2,
            I2=1 / 2,
            Ω=1,
            P=-1,
            electronic_state=states.ElectronicState.B,
        )
    ],
)
```

```python
obe_system = lindblad.setup_OBE_system_transitions(
    trans, transition_selectors, 
    verbose=True,
    qn_compact= True,
    decay_channels = None
)
```
